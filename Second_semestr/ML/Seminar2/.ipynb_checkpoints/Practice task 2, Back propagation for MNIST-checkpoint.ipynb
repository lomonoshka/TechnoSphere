### Часть 1: MNIST
Обучите полносвязную нейронную сеть с архитектурой 784-100-100-10 и сигмоидой в качестве нелинейности. Какую точность классификации удалось получить? Нарисуйте график сходимости на обучающей и тестовой выборках. В качестве темпа обучения (learning rate) возьмите 0.01

```python
error_logs = {0.001: [], 0.005: [], 0.01: []}

for learning_rate in error_logs:
    nn = NeuralNetwork([Linear(784, 100), Sigmoid(), Linear(100,100), Sigmoid(), Linear(100, 10)], NLLLoss())
    for X, y in train_loader:
        X = X.view(X.shape[0], -1)
        X = X.numpy() ### Converts torch.Tensor to numpy array
        y = y.numpy()
        nn.train(X, y, epoch=1, learning_rate=learning_rate)
        error_logs[learning_rate] += [np.mean(nn.log_error)]
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(16, 8))
plt.xlabel('Batch')
plt.ylabel('Mean batch loss')
for learning_rate in error_logs:
    ax.plot(error_logs[learning_rate], label=f'$\lambda$ = {learning_rate}')
ax.legend()
ax.plot();
```

0.01 действительно выглядит лучше своих конкурентов. Посмотрим как уменьшается лосс на тесте при $\lambda = 0.01$

```python
def train_and_view_results(nn, train_loader, test_loader, learning_rate=0.01,
                           plot_error_decrease=True, plot_missclass=True):
    test_loss = []
    for X, y in train_loader:
        X = X.view(X.shape[0], -1)
        X = X.numpy() ### Converts torch.Tensor to numpy array
        y = y.numpy()
        nn.train(X, y, epoch=1, learning_rate=learning_rate, save_prev_log=True)
        X_test, y_test = next(iter(test_loader))
        X_test = X_test.view(X.shape[0], -1)
        X_test = X_test.numpy() ### Converts torch.Tensor to numpy array
        y_test = y_test.numpy()
        res = nn.forward(X_test)
        test_loss.append(nn.loss.forward(res, y_test))
    
    if plot_error_decrease:
        #train loss
        fig, ax = plt.subplots(figsize=(16, 8))
        ax.set_xlabel('Batch')
        ax.set_ylabel('Batch loss')
        ax.plot(nn.log_error, label=f'train loss, $\lambda$ = {learning_rate}')
        ax.plot(test_loss, label=f'test loss, $\lambda$ = {learning_rate}')
        ax.legend()
    
    if plot_missclass:
        X_test = test_dataset.data.numpy()
        X_test = X_test.reshape(X_test.shape[0], -1)
        y_test = test_dataset.targets.numpy()
        y_pred = nn.forward(X_test)
        y_proba = np.exp(y_pred) / np.exp(y_pred).sum(axis=1)[:, None]
        missclass = np.where(np.argmax(y_pred, axis=1) != y_test)[0]
        print(f"accuracy: {1 - len(missclass) / len(y_test)}")
        fig, axes = plt.subplots(2, 5, figsize=(12, 6))
        axes = axes.flatten()
        plt.tight_layout(pad=3)
        for k, i in enumerate(missclass[:10]):
            axes[k].imshow(X_test[i].reshape(28, 28), cmap=plt.cm.Greys_r)
            axes[k].set_title('True label: {}\n'.format(y_test[i]) + \
                      'Predict label {}, p={:.2f}\n'.format(np.argmax(y_proba[i]), np.max(y_proba[i])) + \
                      'True label p={:.2f}'.format(y_proba[i, y_test[i]]))
            axes[k].axis('off')
```

### Sigmoid

```python
nn = NeuralNetwork([Linear(784, 100), Sigmoid(), Linear(100,100), Sigmoid(), Linear(100, 10)], NLLLoss())
train_and_view_results(nn, train_loader, test_loader)
```

### Relu

```python
nn = NeuralNetwork([Linear(784, 100), Relu(), Linear(100,100), Relu(), Linear(100, 10)], NLLLoss())
train_and_view_results(nn, train_loader, test_loader, learning_rate=0.01)
```

### Elu

```python
nn = NeuralNetwork([Linear(784, 100), Elu(), Linear(100,100), Elu(), Linear(100, 10)], NLLLoss())
train_and_view_results(nn, train_loader, test_loader, learning_rate=0.01)
```

### Tanh

```python
nn = NeuralNetwork([Linear(784, 100), Tanh(), Linear(100,100), Tanh(), Linear(100, 10)], NLLLoss())
train_and_view_results(nn, train_loader, test_loader, learning_rate=0.0005)
```

По accuracy самой лучшей моделью оказалась ReLu (accuracy = 0.96). Посмотрим удастся ли улучшить результат если использовать аугментацию
